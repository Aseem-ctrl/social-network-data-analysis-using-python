```python
import json
```

```python
# Load the JSON file

def load_data(filename): #this is a function that takes in a filename as an argument and returns the data from the json file.
    with open(filename, "r") as file: #this is used to open the json file in read mode and store it in a variable called file.
        data = json.load(file)  #this is used to load the data from the json file and store it in a variable called data.
    return data #this is used to return the data that has been loaded from the json file.
```

```python
data = load_data("data.json") #this is used to load the data from the json file and store it in a variable called data.
```

```python
data #this is used to display the data that has been loaded from the json file.
```

```python
# Display users and their connections

def display_users(data):
    print("Users and Their Connections:\n")
    #for user in data["users"]:
    for user in data["users"]: #this is used to loop through the users in the data and display their name, id, friends, and liked pages.
        print(f"{user['name']} (ID: {user['id']}) - Friends: {user['friends']} - Liked Pages: {user['liked_pages']}")

    print("\nPages:\n")
    for page in data["pages"]: #this is used to loop through the pages in the data and display their id and name.
        print(f"{page['id']}: {page['name']}")
```

```python
# Load and display the data

display_users(data)  

#here in data.json the managers cant read and understand the json format data so we convert json data into human readable format so that they can understand the data and make decisions based on it.
```
